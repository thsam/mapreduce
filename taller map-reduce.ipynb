## 2- Índice invertido
Crearemos un índice invertido para una colección de documentos. Un índice invertido nos indica, para cada palabra, en qué documento se encuentra. Análogo al índice de términos que se encuentra al final de algunos libros en donde para cada palabra nos indica en qué páginas aparece.

Inspecciona primero los ficheros books.json, que es la entrada, y solucion.txt, que es la salida que deberia de generar tu programa. ## 4- Amigos asimétricos
En el mismo ejemplo anterior queremos ahora obtener una lista de parejas en las que la relación de amistad es asimétrica. Es decir, si en la entrada tenemos (personA, personB) y también tenemos (personB, personA) entonces ninguna de esas dos tuplas ha de aparecer en la salida. En cambio, si en la entrada tenemos (personA, personB) pero NO TENEMOS (personB, personA), entonces en la salida han de aparecer AMBAS TUPLAS.

Observa los ficheros friends.json y solucion.json para entender lo que se pretende. Observa por ejemplo el caso de Myriel, OldMan que aparecen en la salida y, por otro lado, el caso de Myriel y Valjean que no aparecen en la salida.

Ejecuta tu programa del siguiente modo:

`python friend-assimetric.py friends.json`
